# Reading Images

```python
# getting the path
img_path = os.path.join('data', 'images', 'cat.jpg')
img_path
```

```python
# reading the image
img = cv.imread(img_path)
```

```python
# this is an image represented by a 3-dim matrix
img.dtype, img.shape
#(height, width, channels)
```

```python
# rendering the image (in python)
'''
cv.imshow("Cute_Cat_Display_Window", img) #(window_name, matrix)
k = cv.waitKey(200)
'''
```

```python
# rendering the image in jupyter using matplotlib
plt.imshow(img)
plt.axis(False)
plt.show()
```

One problem with this technique is that, we get the color channels wrong in matplotlib. Matplotlib reads images as RBG, wheras OpenCV reads images as BGR.

```python
# reordering the color channels
# this uses the convert Color method
img_new = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
# reordering the color channels
plt.imshow(img_new)
plt.axis(False)
plt.show()
```

We can also show grayscale version of the images.

```python
# converting to grayscale
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
# rendering the re-colored image
plt.imshow(img_gray)
plt.axis(False)
plt.show()
```

# Writing out Images

```python
op_path = os.path.join('data', 'outputs', 'cat_gray.jpg')
op_path
```

```python
# this writes the image 'cat_gray.jpg' to the specified location
cv.imwrite(op_path, img_gray)
```
